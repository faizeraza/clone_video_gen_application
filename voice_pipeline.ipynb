```python
# create amazon transcribe jobs
import time
import boto3


def transcribe_file(job_name, file_uri, transcribe_client):
    transcribe_client.start_transcription_job(
        TranscriptionJobName=job_name,
        Media={"MediaFileUri": file_uri},
        MediaFormat="wav",
        LanguageCode="en-US",
    )

    max_tries = 60
    while max_tries > 0:
        max_tries -= 1
        job = transcribe_client.get_transcription_job(TranscriptionJobName=job_name)
        job_status = job["TranscriptionJob"]["TranscriptionJobStatus"]
        if job_status in ["COMPLETED", "FAILED"]:
            print(f"Job {job_name} is {job_status}.")
            if job_status == "COMPLETED":
                print(
                    f"Download the transcript from\n"
                    f"\t{job['TranscriptionJob']['Transcript']['TranscriptFileUri']}."
                )
            break
        else:
            print(f"Waiting for {job_name}. Current status is {job_status}.")
        time.sleep(10)


def main():
    transcribe_client = boto3.client("transcribe")
    file_uri = "/content/audio.wav"
    transcribe_file("Example-job", file_uri, transcribe_client)


if __name__ == "__main__":
    main()

```

```python
#hindi replace
!pip install -q googletrans==3.1.0a0
import googletrans
indian_first_names = [
    "Aarav", "Aditi", "Arnav", "Arya", "Diya", "Ishaan", "Kavya", "Riya", "Sohan", "Vanya",
]
names = []
translator = googletrans.Translator()

for i in indian_first_names:
  names.append(translator.translate(i, dest='hi').text)

names
```

```python
#marathi replace
import re
text_arr = []
for i in names:
  text =  'प्रेसिडेंटील स्पीच एचबार पहिलीबार देश केट्रायबल्त शेत्ृसे आदिवासी मैलारको एझे प्रेसिडिन बनाकेई सपोळ पोस्पीरका अर उनकि पहिली प्रेजि डिंडल स्पीचोय  तुमुझे भॉस जादा अपेक्षाती कॉपजीशिनके लोगॉर्तसभी एट पिचरी जातिजमा तीसेच्या वए फी मेल को इचनेबडे पर्पे लाकरमोदिजीने अ अएक रिस्पेक दियतुमुजी लक्त्या ऑपजीशन काबी वहोही कामता कउणकी थाँकिंग स्पीच के बाध उनका थॅकिं करने केबाद फ्लोर्पे आपके जोप प्रश्नते आपकेजो कोय क्वायरीजे वुाआप्रक्सक्तीत ऑपजशन सिर् फंगामा मचारये तु इससे जोलिखाय देरायामुझे लक्तळय वॉपजिशनण ककाम सिरे फंगामामचानय काम करनानये तुतीे जंताखे हीठ्मे काम कर्नेगलीर फोर फ्लोर बनांवय नाकिं आपनी राजनीथ करने केली येफ्लोर बनाय'
  hindi_regex = re.compile('आदिवासी')

# Replace Hindi text with the desired replacement text
 # text = 'नमस्ते'
  replacement_text = i
  replaced_text = re.sub(hindi_regex, replacement_text, text)
  text_arr.append(replaced_text)
# Print the replaced text
text_arr
```
